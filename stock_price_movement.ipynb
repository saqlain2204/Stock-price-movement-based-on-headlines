```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Data.csv", encoding = 'ISO-8859-1')
df.head(3)
```

```python
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
train, test
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot(x='Label', data=train)
```

```python
data = train.iloc[:, 2:27]
data.replace("[^a-zA-Z]", " ", regex = True, inplace = True)
numbers = [i for i in range(25)]
data.columns = numbers
data.head(3)
```

```python
for columns in data.columns:
  data[columns] = data[columns].str.lower()
data.head(3)
```

```python
data['headlines'] = data.apply(lambda row: ' '.join(row.astype(str)), axis=1)
data
```

```python
data["headlines"][0]
```

```python
for column in data.columns:
  if column == 'headlines':
    break
  else:
    data.drop([column], axis = 1, inplace = True)
data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
cv = CountVectorizer(max_features = 2500)
traindataset = cv.fit_transform(data['headlines'])
traindataset
```

```python
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = cv.transform(test_transform)
predictions = randomclassifier.predict(test_dataset)
```

```python
predictions
```

```python
from sklearn.metrics import classification_report, confusion_matrix
report = classification_report(test['Label'], predictions)
print(report)
```

```python
print(confusion_matrix(test['Label'], predictions))
```
